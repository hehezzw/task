```python
import gym
env = gym.make('CartPole-v0')
for i_episode in range(20):
    observation = env.reset()# observation 观察/评测
    # env.reset 用于一个迭代后重启
    for t in range(100):
        ##########
        env.render()
        # 用于每一步后刷新
        ###############
        print(observation)
        ####################
        action = env.action_space.sample()
        print(action)
        ## actions = env.action_space.n() 是代表可以action有多少个返回int值
        ############
        observation, reward, done, info = env.step(action)
        #获取下一步 observation, reward, done, info  是否完成
        #其中observation, reward, done, info 为 环境 得分 检测
        ################
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()





'''
observation  对环境的观察
reward    上一个行动的奖励
done    游戏是否已经结束
info 调试用的诊断信息

env.reset()   重置环境 既重新启动游戏
env.render() 刷新画面
env.action_space.sample()   返回一个action 的随机sample，
也就是说，随机在动作空间中选择一个动作
env.step (action)
特定于环境的对象表示人对环境的观察。
比如 来自摄像机的像素数据 机器人的关节角度和关节速度 或者棋盘中的的棋盘状态
###################
 observation, reward, done, info = env.step(action)
        #获取下一步 observation, reward, done, info  是否完成
        #其中observation, reward, done, info 为 环境 得分 检测
###################

Space 
每个游戏 都有自己的 action_space  and observation_space

'''
```
