```python
import numpy as np
a=np.array([[7,2],[3,4],[5,3]])
print(a)
```

```python
u,d,v = np.linalg.svd(a)
print('u:',u)
print('-----')
print('d:',d)
print('-----')
print('v:',v)
print('-----')
b=np.linalg.eig(a.dot(a.T))[1]
print(b)
```

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
img = cv.imread("d.jpg",0)
print(img)
img = cv.resize(img,(721,833))
print(img)
cv.imshow("a",img)


print(np.min(img))
A = np.array(img)
print(A)
U,D,V = np.linalg.svd(img)
print(U)
print(U,"1")
print(np.linalg.eig(A.dot(A.T))[1],"1.1")
print(V,"2")
print(np.linalg.eig(A.T.dot(A))[1],"2.1")
print(D.size,"3")

for i in [5,10,15,20,30,50]:
    reconstimg = np.matrix(U[:,:i]) * np.diag(D[:i]) * np.matrix(V[:i,:])
    plt.imshow(reconstimg,cmap='gray')
    title = 'n=%s'%i
    plt.title(title)
    plt.show()
```
