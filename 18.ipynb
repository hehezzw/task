```python
a=array([[0,1],[1,1],[2,1],[3,1],[3,1],[4,1]])
b=array([[2],[4],[0],[2],[5],[3]])
c=empty((6,1))
for i in range(6):
    c[i]=a[i][0]
plt.axis([-1,6,-1,6])# the bond limte

plt.title('my dot',fontsize=20,fontname='times new roman')


a_plus=linalg.pinv(a)
print(a_plus)
coefs=a_plus.dot(b)


x_line=linspace(0,5,100)
y_line=coefs[0]*x_line+coefs[1]

plt.grid(True)
plt.plot(c,b,'go')
plt.plot(x_line,y_line)
plt.plot(coefs[0],coefs[1],'bo')
plt.text(coefs[0],coefs[1],'the coefs')

plt.show()
```
